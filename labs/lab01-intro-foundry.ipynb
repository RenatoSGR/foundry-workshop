## O que é o Azure AI Foundry?

O **Azure AI Foundry** é a plataforma unificada da Microsoft para desenvolver aplicações de IA. Combina:

| Componente | Descrição |
|------------|----------|
| **Catálogo de Modelos** | 1600+ modelos (OpenAI, Meta, Mistral, etc.) |
| **Playground** | Interface para testar modelos sem código |
| **Agentes** | Assistentes inteligentes com tools e knowledge |
| **AI Search** | Pesquisa semântica para RAG |
| **Avaliação** | Métricas de qualidade e segurança |
| **Monitoring** | Observabilidade e tracing |

### Arquitetura Simplificada

```
└─────────────────────────────────────────────┘
```

## 🔧 Exercício 1.2: Validar Configuração

Vamos verificar que o ambiente está correctamente configurado.

````python
# Carregar variáveis de ambiente
from dotenv import load_dotenv
import os

load_dotenv("../.env")

# Verificar que todas as variáveis estão definidas
variaveis = [
    "AZURE_AI_FOUNDRY_ENDPOINT",
    "AZURE_AI_FOUNDRY_KEY",
    "AZURE_OPENAI_ENDPOINT",
    "AZURE_OPENAI_KEY",
    "AZURE_OPENAI_DEPLOYMENT",
]

print("=== Verificação do Ambiente ===")
tudo_ok = True
for var in variaveis:
    valor = os.getenv(var)
    if valor and not valor.startswith("<"):
        print(f"  ✅ {var} = {valor[:30]}...")
    else:
        print(f"  ❌ {var} - NÃO CONFIGURADO!")
        tudo_ok = False

if tudo_ok:
    print("\n🎉 Tudo configurado! Podes avançar para o Lab 2.")
else:
    print("\n⚠️  Preenche o ficheiro .env antes de continuar.")
````

````python
# Teste rápido de conectividade ao Azure OpenAI
from openai import AzureOpenAI

client = AzureOpenAI(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_key=os.getenv("AZURE_OPENAI_KEY"),
    api_version=os.getenv("AZURE_OPENAI_API_VERSION", "2025-01-01-preview"),
)

response = client.chat.completions.create(
    model=os.getenv("AZURE_OPENAI_DEPLOYMENT", "gpt-4o"),
    messages=[{"role": "user", "content": "Diz 'Olá Workshop!' em português"}],
    max_tokens=50,
)

print(f"✅ Resposta do modelo: {response.choices[0].message.content}")
print(f"   Modelo: {response.model}")
print(f"   Tokens usados: {response.usage.total_tokens}")
````
